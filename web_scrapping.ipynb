```python
import requests
from bs4 import BeautifulSoup

def get_webpage(url):
    return requests.get(url).text

# Re-defining 'link' here to make the selected cell self-contained and fix the NameError.
webpage_initial = requests.get('https://data.who.int/countries/')
page_initial = BeautifulSoup(webpage_initial.text, 'lxml')

link=[]
for i in page_initial.find_all('div',class_='alphabetical-box'):
  for j in i.find_all('a'):
    link.append(j.attrs['href'])

Name=[]
Population=[]
CHE=[]
WHO_REG=[]
WBIL=[]
counter=1
for l in link:
    page1=BeautifulSoup(get_webpage(l),'lxml')
    Country_box=page1.find_all('div',class_='col-lg-7 col-xxl-6')

    country_data = [] # Use a different variable name to avoid confusion with previous 'data' issues

    # Extract Country Name
    country_name = "N/A"
    if Country_box:
        h1_elements = Country_box[0].find_all('h1')
        if h1_elements:
            country_name = h1_elements[0].text
    country_data.append(country_name)

    # Extract other data from td elements
    if Country_box:
        td_elements = Country_box[0].find_all('td')
        for td_index in range(4): # We expect 4 additional data points
            if td_index < len(td_elements):
                country_data.append(td_elements[td_index].text)
            else:
                country_data.append("N/A") # Fill with N/A if not enough td elements

    # Now check if we collected all 5 expected pieces of data
    if len(country_data) == 5:
        print(counter, country_data[0])
        counter=counter+1
        Name.append(country_data[0])
        Population.append(country_data[1])
        CHE.append(country_data[2])
        WHO_REG.append(country_data[3])
        WBIL.append(country_data[4])
    else:
        # This case should ideally not happen with the current logic of filling with "N/A"
        # but it's a good safeguard.
        print(f"Skipping {l} due to unexpected data length (got {len(country_data)}, expected 5).")
```

```python
import pandas as pd

pd.DataFrame(
    {
    'Country':Name,
    'Population':Population,
    "Current Health Expenditure":CHE,
    'WHO REGION':WHO_REG,
    'World Bank income level':WBIL
    }
).to_csv('Data.csv')
```
